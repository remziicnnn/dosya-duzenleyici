# ðŸ“ Dosya DÃ¼zenleyici (Python)

Bu notebook, verilen bir klasÃ¶rdeki dosyalarÄ± uzantÄ±larÄ±na gÃ¶re
otomatik olarak alt klasÃ¶rlere ayÄ±rÄ±r.

## Ã–zellikler
- KullanÄ±cÄ±dan klasÃ¶r yolu alÄ±r
- Deneme (dry-run) ve gerÃ§ek taÅŸÄ±ma modu vardÄ±r
- DosyalarÄ± uzantÄ±larÄ±na gÃ¶re sÄ±nÄ±flandÄ±rÄ±r

âš ï¸ Ã–neri: Ä°lk Ã§alÄ±ÅŸtÄ±rmada **Deneme modu** kullanÄ±n.

```python
import os
import shutil


def dosyalari_duzenle():
    # KullanÄ±cÄ±dan klasÃ¶r yolu al
    klasor_yolu = input("DÃ¼zenlenecek klasÃ¶r yolunu gir: ").strip()

    if not os.path.isdir(klasor_yolu):
        print("âŒ GeÃ§erli bir klasÃ¶r yolu girmedin.")
        return

    print("\nNasÄ±l Ã§alÄ±ÅŸtÄ±rÄ±lsÄ±n?")
    print("1 - Deneme (dosyalar taÅŸÄ±nmaz)")
    print("2 - GerÃ§ek taÅŸÄ±ma")

    secim = input("SeÃ§imin (1/2): ").strip()

    if secim == "1":
        dry_run = True
    elif secim == "2":
        dry_run = False
    else:
        print("GeÃ§ersiz seÃ§im.")
        return

    # UzantÄ± -> KlasÃ¶r eÅŸleÅŸmesi
    uzanti_klasor_map = {
        ".jpg": "IMAGES",
        ".jpeg": "IMAGES",
        ".png": "IMAGES",
        ".gif": "IMAGES",

        ".pdf": "PDF",
        ".doc": "DOCS",
        ".docx": "DOCS",
        ".txt": "TEXT",

        ".mp3": "MUSIC",
        ".wav": "MUSIC",

        ".mp4": "VIDEOS",
        ".mkv": "VIDEOS",

        ".py": "PYTHON",
        ".ipynb": "NOTEBOOKS"
    }

    tasinan_sayisi = 0

    for eleman in os.listdir(klasor_yolu):
        tam_yol = os.path.join(klasor_yolu, eleman)

        if os.path.isdir(tam_yol):
            continue

        _, ext = os.path.splitext(eleman)
        ext = ext.lower()

        hedef_klasor = uzanti_klasor_map.get(ext, "OTHERS")
        hedef_klasor_yolu = os.path.join(klasor_yolu, hedef_klasor)

        print(f"{eleman} â†’ {hedef_klasor}")

        if dry_run:
            continue

        os.makedirs(hedef_klasor_yolu, exist_ok=True)
        shutil.move(tam_yol, os.path.join(hedef_klasor_yolu, eleman))
        tasinan_sayisi += 1

    if dry_run:
        print("\nðŸ§ª Deneme tamamlandÄ±. HiÃ§bir dosya taÅŸÄ±nmadÄ±.")
    else:
        print(f"\nâœ… {tasinan_sayisi} dosya baÅŸarÄ±yla dÃ¼zenlendi.")
```

```python
dosyalari_duzenle()
```
